```python
import pandas as pd
```

```python
arquivo_globo = r'Dados/Historico_de_materias.csv'
df_globo = pd.read_csv(arquivo_globo)
df_globo.head()
```

```python
arquivo_folha = r'Dados/articles.csv'
df_folha = pd.read_csv(arquivo_folha)
df_folha
```

```python
globo_agrupado = df_globo.groupby('data')['titulo'].apply(lambda x: '. '.join(x)).reset_index()
globo_agrupado
```

```python
folha_agrupado = df_folha.groupby('date')['title'].apply(lambda x: '. '.join(x)).reset_index()
folha_agrupado['date'] = pd.to_datetime(folha_agrupado['date'])
folha_agrupado = folha_agrupado.set_index(folha_agrupado['date'])
del folha_agrupado['date']
folha_agrupado
```

```python
print('Folha:', folha_agrupado.index.max(), folha_agrupado.index.min()) 
print('Globo:', globo_agrupado.data.max(), globo_agrupado.data.min())
```

```python
start_date = folha_agrupado.index.min()
end_date = folha_agrupado.index.max()
todos_dias = pd.date_range(start=start_date, end=end_date)
folha_agrupado_ajustado = folha_agrupado.reindex(todos_dias)
display(folha_agrupado_ajustado)
```
